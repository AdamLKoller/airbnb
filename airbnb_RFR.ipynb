### Importing modules

```python
from airbnb_preperation import airbnb_preperation
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
import numpy as np
```

### Importing the prepared data

```python
Xtrain, ytrain, Xdev, ydev, Xtest, ytest = airbnb_preperation('./Data')
```

### Grid search for best hyperparameters (Random Forest Regression)

```python
def grid_search_hyper_parmas(X_train, y_train):
    n_estimators = [int(x) for x in np.linspace(start = 10, stop = 200, num = 4)]
    max_features = ['sqrt','log2']
    max_depth = [int(x) for x in np.linspace(10, 110, num = 6)]
    max_depth.append(None)
    min_samples_split = [2, 5, 10]
    min_samples_leaf = [1, 2, 4]
    bootstrap = [True, False]


    param_grid = {'n_estimators': n_estimators,
                  'max_features':max_features,
                  'max_depth':max_depth,
                  'min_samples_split':min_samples_split,
                  'min_samples_leaf':min_samples_leaf,
                  'bootstrap':bootstrap}
    
    print('Training...')
    grid= GridSearchCV(RandomForestRegressor(),
                       param_grid,
                       return_train_score = True,
                       refit = True,
                       verbose =3,
                       n_jobs=1,
                       cv=3)
    grid.fit(X_train, y_train)
    print("Grid searching is done!")
    print("The best score: ", grid.best_score_)
    print("The best hyperparameter:")
    print(grid.best_params_)
    return grid
```

```python
%%time
grid_search_hyper_parmas(Xtrain, ytrain)
```
